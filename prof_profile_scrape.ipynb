```python
import requests
from bs4 import BeautifulSoup
```

```python
import re
```

```python
url = "https://www.uhd.edu/academics/sciences/computer-science-engineering-technology/Pages/directory.aspx"
```

```python
request = requests.get(url)
content = request.content
```

```python
parser = BeautifulSoup(content, "lxml")
```

```python
tbody = parser.find_all("table")
```

```python
tables = []
```

```python
for tbody in parser.select("tbody"):
    tables.append(tbody)
    print(tbody.text)
    print("\n")
```

```python
department_chairs = tables[0]
comp_sci = tables[1]
engineer_tech = tables[2]
adjunct = tables[3]
staff = tables[4]
```

```python
def get_content(url):
    request = requests.get(url)
    content = request.content

    parser = BeautifulSoup(content, "lxml")

    biography = parser.select("#biography")[0]
    interests = parser.select("#interests")[0]
    courses = parser.select("#courses")[0]
    publications = parser.select("#publications")[0]
    awards = parser.select("#awards")[0]
    
    profile_tabs = [biography, interests, courses, publications, awards]
    
    for tabs in profile_tabs:
        for info in tabs.find_all(["p", "li"]):
            print(info)
            print("\n")
            print("************************")
```

```python
for tr in department_chairs.find_all('tr', class_= re.compile(r'ms-rteTable(Odd)|(Even)')):
    
    profile_url = tr.find('a', href=re.compile(r'(.aspx)$'))
    profile_url = profile_url['href']
    
    uhd_url = "https://www.uhd.edu"
    url = uhd_url + profile_url
    
    get_content(url)
    
    print("\n")
```
